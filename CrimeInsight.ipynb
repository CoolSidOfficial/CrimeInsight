```python
import pandas as pd
```

```python
df=pd.read_csv("processed_file.csv")
```

```python
df
```

```python
df.drop(columns="Time of Occurrence",inplace=True)
```

```python
df
```

```python
target=df[["Crime Description"]]
```

```python
features=df.drop(columns="Crime Description")
```

```python
target
```

```python
features
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_city = LabelEncoder()
le_gender=LabelEncoder()
le_desc=LabelEncoder()
```

```python
features['City'] = le_city.fit_transform(features['City'])
features['Victim Gender'] = le_gender.fit_transform(features['Victim Gender'])
```

```python
target["Crime Description"]=le_desc.fit_transform(target["Crime Description"])
```

```python
target
```

```python
target
```

```python
features
```

```python
features['Time'] = features['Time'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(features,target)
```
